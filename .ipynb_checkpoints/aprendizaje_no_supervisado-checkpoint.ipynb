```python
# Importar las bibliotecas necesarias para este proyecto
import numpy as np  # Para operaciones matemáticas y manejo de arrays
import pandas as pd  # Para manipulación y análisis de datos
from IPython.display import display  # Permite usar display() para mostrar DataFrames

# Bibliotecas de scikit-learn para machine learning
from sklearn.model_selection import train_test_split  # Para dividir datos en conjuntos
from sklearn.tree import DecisionTreeRegressor  # Para árboles de decisión
from sklearn.decomposition import PCA  # Para análisis de componentes principales
from sklearn.cluster import KMeans  # Para algoritmo de clustering K-Means
from sklearn.metrics import silhouette_score  # Para evaluar la calidad de los clusters

# Bibliotecas para visualización de datos
import seaborn as sns  # Para gráficos estadísticos
import matplotlib.pyplot as plt  # Para crear gráficos

# Configuración para mostrar gráficos en el notebook
%matplotlib inline

# Cargar el conjunto de datos de clientes mayoristas
try:
    # Leer el archivo CSV con los datos de clientes
    data = pd.read_csv("./Wholesale customers data.csv")
    
    # Eliminar las columnas 'Region' y 'Channel'
    # axis=1 indica columnas, inplace=True modifica el DataFrame original
    data.drop(['Region', 'Channel'], axis=1, inplace=True)
    
    # Mostrar información sobre el tamaño del conjunto de datos
    print("El conjunto de datos de clientes mayoristas tiene {} muestras con {} características cada una.".format(*data.shape))
    
except:
    # Mensaje de error si no se puede cargar el archivo
    print("No se pudo cargar el conjunto de datos. ¿Falta el archivo de datos?")
```

```python
###########################################
# Suprimir advertencias de usuario de matplotlib
# Necesario para versiones más nuevas de matplotlib
import warnings
warnings.filterwarnings("ignore", category = UserWarning, module = "matplotlib")
#
# Mostrar gráficos de matplotlib en línea con IPython
from IPython import get_ipython
get_ipython().run_line_magic('matplotlib', 'inline')
###########################################

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import pandas as pd
import numpy as np

def pca_results(good_data, pca):
	'''
	Crear un DataFrame de los resultados de PCA
	Incluye pesos de características de dimensiones y varianza explicada
	Visualiza los resultados del PCA
	'''

	# Indexación de dimensiones
	dimensions = dimensions = ['Dimension {}'.format(i) for i in range(1,len(pca.components_)+1)]

	# Componentes del PCA
	components = pd.DataFrame(np.round(pca.components_, 4), columns = list(good_data.keys()))
	components.index = dimensions

	# Varianza explicada del PCA
	ratios = pca.explained_variance_ratio_.reshape(len(pca.components_), 1)
	variance_ratios = pd.DataFrame(np.round(ratios, 4), columns = ['Varianza Explicada'])
	variance_ratios.index = dimensions

	# Crear una visualización de gráfico de barras
	fig, ax = plt.subplots(figsize = (14,8))

	# Graficar los pesos de las características como función de los componentes
	components.plot(ax = ax, kind = 'bar');
	ax.set_ylabel("Pesos de Características")
	ax.set_xticklabels(dimensions, rotation=0)

	# Mostrar las proporciones de varianza explicada
	for i, ev in enumerate(pca.explained_variance_ratio_):
		ax.text(i-0.40, ax.get_ylim()[1] + 0.05, "Varianza Explicada\n          %.4f"%(ev))

	# Devolver un DataFrame concatenado
	return pd.concat([variance_ratios, components], axis = 1)

def cluster_results(reduced_data, preds, centers, pca_samples):
	'''
	Visualiza los datos de clusters reducidos por PCA en dos dimensiones
	Agrega pistas para centros de clusters y datos de muestra seleccionados
	'''

	predictions = pd.DataFrame(preds, columns = ['Cluster'])
	plot_data = pd.concat([predictions, reduced_data], axis = 1)

	# Generar el gráfico de clusters
	fig, ax = plt.subplots(figsize = (14,8))

	# Mapa de colores
	cmap = cm.get_cmap('gist_rainbow')

	# Colorear los puntos basado en el cluster asignado
	for i, cluster in plot_data.groupby('Cluster'):   
	    cluster.plot(ax = ax, kind = 'scatter', x = 'Dimension 1', y = 'Dimension 2', \
	                 color = cmap((i)*1.0/(len(centers)-1)), label = 'Cluster %i'%(i), s=30);

	# Graficar centros con indicadores
	for i, c in enumerate(centers):
	    ax.scatter(x = c[0], y = c[1], color = 'white', edgecolors = 'black', \
	               alpha = 1, linewidth = 2, marker = 'o', s=200);
	    ax.scatter(x = c[0], y = c[1], marker='$%d$'%(i), alpha = 1, s=100);

	# Graficar puntos de muestra transformados
	ax.scatter(x = pca_samples[:,0], y = pca_samples[:,1], \
	           s = 150, linewidth = 4, color = 'black', marker = 'x');

	# Establecer título del gráfico
	ax.set_title("Aprendizaje de Clusters en Datos Reducidos por PCA - Centroides Marcados por Número\nDatos de Muestra Transformados Marcados por Cruz Negra");


def biplot(good_data, reduced_data, pca):
    '''
    Producir un biplot que muestra un diagrama de dispersión de los datos reducidos
    y las proyecciones de las características originales.
    
    good_data: datos originales, antes de la transformación.
               Necesita ser un pandas dataframe con nombres de columnas válidos
    reduced_data: los datos reducidos (las primeras dos dimensiones se grafican)
    pca: objeto pca que contiene el atributo components_

    return: un objeto matplotlib AxesSubplot (para cualquier personalización adicional)
    
    Este procedimiento está inspirado en el script:
    https://github.com/teddyroland/python-biplot
    '''

    fig, ax = plt.subplots(figsize = (14,8))
    # diagrama de dispersión de los datos reducidos
    ax.scatter(x=reduced_data.loc[:, 'Dimension 1'], y=reduced_data.loc[:, 'Dimension 2'], 
        facecolors='b', edgecolors='b', s=70, alpha=0.5)
    
    feature_vectors = pca.components_.T

    # usamos factores de escala para hacer las flechas más fáciles de ver
    arrow_size, text_pos = 7.0, 8.0,

    # proyecciones de las características originales
    for i, v in enumerate(feature_vectors):
        ax.arrow(0, 0, arrow_size*v[0], arrow_size*v[1], 
                  head_width=0.2, head_length=0.2, linewidth=2, color='red')
        ax.text(v[0]*text_pos, v[1]*text_pos, good_data.columns[i], color='black', 
                 ha='center', va='center', fontsize=18)

    ax.set_xlabel("Dimension 1", fontsize=14)
    ax.set_ylabel("Dimension 2", fontsize=14)
    ax.set_title("Plano PC con proyecciones de características originales.", fontsize=16);
    return ax
    

def channel_results(reduced_data, outliers, pca_samples):
	'''
	Visualiza los datos de clusters reducidos por PCA en dos dimensiones usando el conjunto de datos completo
	Los datos están etiquetados por "Channel" y se agregan pistas para datos de muestra seleccionados
	'''

	# Verificar que el conjunto de datos se pueda cargar
	try:
	    full_data = pd.read_csv("../input/customers.csv")
	except:
	    print("No se pudo cargar el conjunto de datos. ¿Falta el archivo?")       
	    return False

	# Crear el DataFrame de Channel
	channel = pd.DataFrame(full_data['Channel'], columns = ['Channel'])
	channel = channel.drop(channel.index[outliers]).reset_index(drop = True)
	labeled = pd.concat([reduced_data, channel], axis = 1)
	
	# Generar el gráfico de clusters
	fig, ax = plt.subplots(figsize = (14,8))

	# Mapa de colores
	cmap = cm.get_cmap('gist_rainbow')

	# Colorear los puntos basado en el Channel asignado
	labels = ['Hotel/Restaurante/Café', 'Minorista']
	grouped = labeled.groupby('Channel')
	for i, channel in grouped:   
	    channel.plot(ax = ax, kind = 'scatter', x = 'Dimension 1', y = 'Dimension 2', \
	                 color = cmap((i-1)*1.0/2), label = labels[i-1], s=30);
	    
	# Graficar puntos de muestra transformados
	for i, sample in enumerate(pca_samples):
		ax.scatter(x = sample[0], y = sample[1], \
	           s = 200, linewidth = 3, color = 'black', marker = 'o', facecolors = 'none');
		ax.scatter(x = sample[0]+0.25, y = sample[1]+0.3, marker='$%d$'%(i), alpha = 1, s=125);

	# Establecer título del gráfico
	ax.set_title("Datos Reducidos por PCA Etiquetados por 'Channel'\nDatos de Muestra Transformados Encerrados en Círculo");
```

#### Implementación: Relevancia de las características
Una reflexión interesante es si una (o más) de las seis categorías de productos es realmente relevante para comprender las compras de los clientes. Es decir, ¿es posible determinar si los clientes que compran cierta cantidad de una categoría de productos necesariamente comprarán una cantidad proporcional de otra? Podemos determinarlo fácilmente entrenando un modelo de aprendizaje de regresión supervisada con un subconjunto de datos sin una característica y evaluando la capacidad de ese modelo para predecir dicha característica.

En el bloque de código a continuación, deberá implementar lo siguiente:

°Asigne new_datauna copia de los datos eliminando una característica de su elección mediante la DataFrame.dropfunción.

°Úselo sklearn.cross_validation.train_test_splitpara dividir el conjunto de datos en conjuntos de entrenamiento y prueba.

    °Utilice la función eliminada como etiqueta de destino. Establezca un test_sizede 0.25y un random_state.

°Importe un regresor de árbol de decisión, establezca un random_statey ajuste el alumno a los datos de entrenamiento.

°Informe la puntuación de predicción del conjunto de prueba utilizando la scorefunción del regresor.

```python
def predict_one_feature(dropped_feature):
    # Hacer una copia del DataFrame, usando la función 'drop' para eliminar la característica dada
    print("Eliminando característica -> {}".format(dropped_feature))
    new_data = data.drop([dropped_feature], axis=1)
    
    # Dividir los datos en conjuntos de entrenamiento y prueba (25%) usando la característica eliminada como objetivo
    # Establecer una semilla aleatoria (random state)
    X_train, X_test, y_train, y_test = train_test_split(new_data, data[dropped_feature], test_size=0.25, random_state=0)
    
    # Crear un regresor de árbol de decisión y ajustarlo al conjunto de entrenamiento
    regressor = DecisionTreeRegressor(random_state=0)
    regressor.fit(X_train, y_train)
    
    # Mostrar el puntaje de la predicción usando el conjunto de prueba
    score = regressor.score(X_test, y_test)
    print("Puntaje al predecir '{}' usando las otras características = {:.3f}\n".format(dropped_feature, score))
```

```python
# Intento de predecir el puntaje de 'Milk' usando las otras características
predict_one_feature('Milk')
```

```python
# Mostrar las características (columnas) presentes en el conjunto de datos
print("Características en los datos -> {}\n".format(data.columns.values))

# Predecir el puntaje de cada característica eliminándola y usando las demás
for cols in data.columns.values:
    predict_one_feature(cols)
```

#### Preprocesamiento de datos

En esta sección, preprocesaremos los datos para crear una mejor representación de los clientes mediante un escalado de los datos y la detección (y, opcionalmente, la eliminación) de valores atípicos. El preprocesamiento de datos suele ser un paso crucial para garantizar que los resultados obtenidos del análisis sean significativos.

#### Implementación: Escalado de características
Si los datos no se distribuyen normalmente, especialmente si la media y la mediana varían significativamente (lo que indica una gran asimetría), suele ser adecuado aplicar un escalamiento no lineal, sobre todo para datos financieros. Una forma de lograr este escalamiento es mediante la prueba de Box-Cox , que calcula la mejor transformación de potencia de los datos que reduce la asimetría. # Implementación: Detección de valores atípicos
Detectar valores atípicos en los datos es fundamental durante el preprocesamiento de cualquier análisis. La presencia de valores atípicos puede sesgar los resultados que los consideran. Existen diversas reglas generales para determinar qué constituye un valor atípico en un conjunto de datos. En este caso, utilizaremos el método de Tukey para identificar valores atípicos : un paso atípico se calcula como 1,5 veces el rango intercuartil (RIC). Un punto de datos con una característica que supera un paso atípico fuera del RIC para esa característica se considera anormal.

En el bloque de código a continuación, deberá implementar lo siguiente:

°Asigna el valor del percentil 25 para la característica dada a Q1. Usa np.percentilepara esto.

°Asigna el valor del percentil 75 para la característica dada a Q3. Nuevamente, usa np.percentile.

°Asigna el cálculo de un paso de valor atípico para la característica dada a step.

°Opcionalmente, puede eliminar puntos de datos del conjunto de datos agregando índices a la outlierslista.

NOTA: Si decide eliminar valores atípicos, asegúrese de que los datos de muestra no contengan ninguno de estos puntos.
Una vez realizada esta implementación, el conjunto de datos se almacenará en la variable good_data.

```python
outliers_list = []
# Para cada característica, encontrar los puntos de datos con valores extremos altos o bajos
for feature in log_data.keys():
    
    # Calcular Q1 (percentil 25 de los datos) para la característica dada
    Q1 = np.percentile(log_data[feature], 25)
    
    # Calcular Q3 (percentil 75 de los datos) para la característica dada
    Q3 = np.percentile(log_data[feature], 75)
    
    # Usar el rango intercuartílico para calcular el paso de outlier (1.5 veces el rango intercuartílico)
    step = (Q3 - Q1) * 1.5
    
    # Mostrar los valores atípicos (outliers)
    print("Puntos de datos considerados outliers para la característica '{}':".format(feature))
    outliers = list(log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))].index.values)
    display(log_data[~((log_data[feature] >= Q1 - step) & (log_data[feature] <= Q3 + step))])
    outliers_list.extend(outliers)
    
print("Lista de outliers -> {}".format(outliers_list))
duplicate_outliers_list = list(set([x for x in outliers_list if outliers_list.count(x) >= 2]))
duplicate_outliers_list.sort()
print("\nLista de outliers comunes -> {}".format(duplicate_outliers_list))

# Seleccionar los índices de los puntos de datos que se desean eliminar
outliers  = duplicate_outliers_list

# Eliminar los outliers, si se especificaron algunos
good_data = log_data.drop(log_data.index[outliers]).reset_index(drop = True)
```

#### Transformación de características
En esta sección, utilizará el análisis de componentes principales (ACP) para extraer conclusiones sobre la estructura subyacente de los datos de clientes mayoristas. Dado que el ACP en un conjunto de datos calcula las dimensiones que maximizan la varianza, determinaremos qué combinaciones de características compuestas describen mejor a los clientes.

#### Implementación: PCA
Ahora que los datos se han escalado a una distribución más normal y se han eliminado los valores atípicos necesarios, podemos aplicar el ACP para good_datadescubrir qué dimensiones de los datos maximizan mejor la varianza de las características involucradas. Además de encontrar estas dimensiones, el ACP también reportará la razón de varianza explicada de cada dimensión: cuánta varianza dentro de los datos se explica únicamente por esa dimensión. Cabe destacar que un componente (dimensión) del ACP puede considerarse una nueva "característica" del espacio; sin embargo, es una composición de las características originales presentes en los datos.

En el bloque de código a continuación, deberá implementar lo siguiente:

°Importar sklearn.decomposition.PCAy asignar los resultados del ajuste de PCA en seis dimensiones con good_dataa pca.

°Aplicar una transformación PCA de log_samplesutilizando pca.transformy asignar los resultados a pca_samples.

```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Apply PCA by fitting the good data with the same number of dimensions as features
pca = PCA(n_components=6, random_state=0)
pca.fit(good_data)

# Transform log_samples using the PCA fit above
pca_samples = pca.transform(log_samples)

print("Explained Variance Ratio => {}\n".format(pca.explained_variance_ratio_))
print("Explained Variance Ratio(csum) => {}\n".format(pca.explained_variance_ratio_.cumsum()))

# Generate PCA results plot
def create_pca_plot(good_data, pca):
    # Create the components dataframe
    dimensions = ['PC{}'.format(i+1) for i in range(len(pca.components_))]
    components = pd.DataFrame(np.round(pca.components_, 4), columns=good_data.keys())
    components.index = dimensions
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Create heatmap-style bar plot for each component
    x_pos = np.arange(len(dimensions))
    width = 0.13
    
    colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
    
    for i, feature in enumerate(good_data.columns):
        values = components[feature].values
        ax.bar(x_pos + i*width, values, width, label=feature, color=colors[i])
    
    ax.set_xlabel('Principal Components')
    ax.set_ylabel('Feature Weight')
    ax.set_title('PCA Feature Weights by Component\nExplained Variance: ' + 
                 ', '.join([f'{var:.4f}' for var in pca.explained_variance_ratio_]))
    ax.set_xticks(x_pos + width * 2.5)
    ax.set_xticklabels(dimensions)
    ax.legend()
    ax.grid(True, alpha=0.3)
    
    # Add explained variance text on top
    for i, (dim, var) in enumerate(zip(dimensions, pca.explained_variance_ratio_)):
        ax.text(i, ax.get_ylim()[1]*0.9, f'Explained Variance\n{var:.4f}', 
                ha='center', va='top', fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat'))
    
    plt.tight_layout()
    plt.show()
    
    return components

# Call the function to create the plot
pca_results = create_pca_plot(good_data, pca)
```

# Implementación: Recuperación de datos
Cada clúster presente en la visualización anterior tiene un punto central. Estos centros (o medias) no son puntos de datos específicos, sino los promedios de todos los puntos de datos predichos en los respectivos clústeres. Para la creación de segmentos de clientes, el punto central de un clúster corresponde al cliente promedio de ese segmento . Dado que los datos tienen una dimensión reducida y una escala logarítmica, podemos recuperar el gasto representativo del cliente a partir de estos puntos de datos aplicando las transformaciones inversas.

En el bloque de código a continuación, deberá implementar lo siguiente:

    Aplique la transformada inversa al centersuso pca.inverse_transformy asigne los nuevos centros a log_centers.

    Aplicar la función inversa de np.loga log_centersutilizando np.expy asignar los centros verdaderos a true_centers.

```python
# Inverse transform the centers
log_centers = pca.inverse_transform(centers)

# Exponentiate the centers
true_centers = np.exp(log_centers)

# Display the true centers
segments = ['Segment {}'.format(i) for i in range(0,len(centers))]
true_centers = pd.DataFrame(np.round(true_centers), columns = data.keys())
true_centers.index = segments
display(true_centers)
```
